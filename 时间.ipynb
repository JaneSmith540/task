```python
import time
from Data_Handling import get_price
import pandas as pd

def measure_get_price_time(security, count=1, fields=['close'], end_date=None):
    """
    测量 get_price 函数的执行时间
    
    Parameters:
    security: 股票代码
    count: 获取数据的数量
    fields: 获取的字段列表
    end_date: 结束日期
    
    Returns:
    execution_time: 执行时间（秒）
    data: 获取到的数据
    """
    if end_date is None:
        end_date = pd.Timestamp.now()
    
    # 记录开始时间
    start_time = time.time()
    
    # 执行 get_price 函数
    data = get_price(security=['300450.SZ'], start_date='2020-01-01', end_date='2020-12-31', fields=['open', 'close', 'vol'])
    
    # 记录结束时间
    end_time = time.time()
    
    # 计算执行时间
    execution_time = end_time - start_time
    
    print(f"get_price 执行时间: {execution_time:.4f} 秒")
    print(f"获取到 {len(data)} 条数据")
    
    return execution_time, data

# 使用示例
# execution_time, data = measure_get_price_time('600612.SH')
```

```python
import time

def measure_print_time(*args, **kwargs):
    """
    测量 print 函数的执行时间
    
    Parameters:
    *args: 传递给 print 的参数
    **kwargs: 传递给 print 的关键字参数
    
    Returns:
    execution_time: 执行时间（秒）
    """
    # 记录开始时间
    start_time = time.perf_counter()
    
    # 执行 print 函数
    print(*args, **kwargs)
    
    # 记录结束时间
    end_time = time.perf_counter()
    
    # 计算执行时间
    execution_time = end_time - start_time
    
    return execution_time

# 使用示例
# execution_time = measure_print_time("Hello, World!")
# print(f"print 函数执行时间: {execution_time:.6f} 秒")
```
